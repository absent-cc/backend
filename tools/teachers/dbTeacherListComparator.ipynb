```python
import csv

PATHS = {
    "NEWTON_SOUTH": "/home/kevin/projects/abSENT/backend/data/NSHS_teachers.csv",
    "NEWTON_NORTH": "/home/kevin/projects/abSENT/backend/data/NNHS_teachers.csv",
}
```

```python
teachers_names = {
    "NEWTON_SOUTH": [],
    "NEWTON_NORTH": [],
}

for school, path in PATHS.items():
    with open(path, "r") as f:
        reader = csv.reader(f)
        for row in reader:
            first_name = row[0]
            last_name = row[1]
            teachers_names[school].append((first_name.lower(), last_name.lower()))

with open("/home/kevin/projects/abSENT/backend/data/teachers.csv", "r") as f:
    teachers = csv.DictReader(f)
    for row in teachers:
        school = row["school"]
        first = row["first"]
        last = row["last"]
        if (first.lower(), last.lower()) not in teachers_names[school]:
            print(school, first, last)
```
