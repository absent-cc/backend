```python
import csv

from typing import List, Tuple
```

```python
faculty_path = 'tools/teachers/schoologyFaculty/NSHS_faculty.csv'
current_path = 'data_backup/NSHS_teachers.csv'
```

```python
with open("data/NNHS_faculty.csv", "r") as f:
    data = csv.DictReader(f)
    FACULTY_NAMES: List[Tuple[str, str]] = []
    for col in data:
        FACULTY_NAMES.append((col["First"], col["Last"]))
```

```python
with open("data/NNHS_teachers.csv", "r") as f:
    data = csv.DictReader(f)
    TEACHER_NAMES: List[Tuple[str, str]] = []
    for col in data:
        TEACHER_NAMES.append((col["First"], col["Last"]))
```

```python
teacher_set = set(TEACHER_NAMES)
faculty_set = set(FACULTY_NAMES)
```

```python
difference = faculty_set - teacher_set
print(difference)
```

```python
with open('data/NNHS_teachers.csv', "a") as f:
    writer = csv.writer(f)
    for teacher in difference:
        teacher_to_add = list(teacher)
        print(teacher_to_add)
        writer.writerow(teacher_to_add)
```

```python
with open("data/NSHS_teachers.csv", "r") as f:
    data = csv.DictReader(f)
    NEW_TEACHERS: List[Tuple[str, str]] = []
    for col in data:
        NEW_TEACHERS.append((col["First"], col["Last"]))
```

```python
sorted_new_teachers = sorted(NEW_TEACHERS)
```

```python
print(sorted_new_teachers)
```

```python
with open('data/NSHS_teachers_final.csv', "w+") as f:
    writer = csv.writer(f)
    for teacher in sorted_new_teachers:
        teacher_to_add = list(teacher)
        print(teacher_to_add)
        writer.writerow(teacher_to_add)
```
