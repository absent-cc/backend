```python
# Imports
import csv
from typing import List, Tuple
from fuzzywuzzy import fuzz
```

```python
with open("data/NNHS_faculty.csv", "r") as f:
    data = csv.DictReader(f)
    FACULTY_NAMES: List[Tuple[str, str]] = []
    for col in data:
        FACULTY_NAMES.append((col["First"], col["Last"]))
```

```python
with open("data/NNHS_teachers.csv", "r") as f:
    data = csv.DictReader(f)
    TEACHER_NAMES: List[Tuple[str, str]] = []
    for col in data:
        TEACHER_NAMES.append((col["First"], col["Last"]))
```

```python
def fuzzySearch(name: Tuple[str, str], names: List[Tuple[str, str]]):
    first_compare = name[0]
    last_compare = name[1]
    
    for first, last in names:
        first_fuzz_amount = fuzz.ratio(first_compare.lower(), first.lower())
        last_fuzz_amount = fuzz.ratio(last_compare.lower(), last.lower())

        first_criteria = first_fuzz_amount > FUZZ_THRESHOLD and first_fuzz_amount != 100
        last_criteria = last_fuzz_amount > FUZZ_THRESHOLD and last_fuzz_amount != 100

        if first_criteria and last_criteria:
            print(f"{first_compare} {last_compare} : {first} {last}")
    
        print(f"{first_compare} {first_fuzz_amount}, {first}")
```
