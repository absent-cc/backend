```python
FUZZ_THRESHOLD = 140

def fuzzySearch(name: Tuple[str, str], names: List[Tuple[str, str]]):
    firstCompare = name[0]
    lastCompare = name[1]

    for first, last in names:
        first_fuzz = fuzz.ratio(firstCompare.lower(), first.lower())
        last_fuzz = fuzz.ratio(lastCompare.lower(), last.lower())

        full_name_fuzz = first_fuzz + last_fuzz
        full_name_criteria = full_name_fuzz > FUZZ_THRESHOLD and full_name_fuzz < 185

        # first_criteria = first_fuzz > FUZZ_THRESHOLD and first_fuzz != 100
        last_criteria = last_fuzz > 35 and last_fuzz != 100

        if full_name_criteria and last_criteria:
            print(f"{first} {last} : {name} {full_name_fuzz}, {first_fuzz}, {last_fuzz}")
```
